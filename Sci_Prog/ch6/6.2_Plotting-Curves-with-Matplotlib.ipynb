# 6.2 Plotting Curves with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
x = np.linspace(0, 10, n+1)
y = np.exp(-x) * np.sin(2*np.pi*x)
# print(x)
# print(y)

plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return np.exp(-x)*np.sin(2*np.pi*x)

n = 100
x = np.linspace(0, 8, n+1)
y = f(x)

# print(x)
# print(y)

# plt.plot(x, y, label = "exp(-x)*sin(2$\pi$ x)")
plt.plot(x, y, label = "exp(-x)*sin(2$\pi$x)")

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.axis([-0.5, 8, -0.5, 0.8])
plt.title("My first Matplotlib Demo Chart")

plt.savefig("fig.pdf")
plt.savefig("fig.jpg")
plt.savefig("fig.png")
plt.savefig("fig.svg")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f1(x):
    return np.exp(-x)*np.sin(2*np.pi*x)

def f2(x):
    return np.exp(-2*x)*np.sin(4*np.pi*x)

x = np.linspace(0, 8, 401)
y1 = f1(x)
y2 = f2(x)

plt.plot(x, y1, 'r--', label="exp(-x)*sin(2$\pi$x)")
plt.plot(x, y2, 'b:', label="exp(-2x)*sin(4$\pi$x)")
plt.title("Plotting two curves in the same plot")
plt.xlabel("x")
plt.ylabel("y")

plt.legend()

plt.savefig("Figure_Two_Curve.png")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 8, 201)
plt.plot(
    x,
    np.exp(-x)*np.sin(2*np.pi*x),
    x,
    np.exp(-2*x)*np.sin(4*np.pi*x)
)
plt.show()
```
