# 6.3 Plotting Discontinuous and Piecewise Functions

```python
def H(x):
    if x < 0:
        return 0
    else:
        return 1

print(H(-5), H(10))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 5)
print(x, type(x))
# print(b = x < 0)
y = H(x)
print(type(y))
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 5)
print(x, type(x))
print(b = x < 0)
y = H(x)
print(type(y))
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

n = 10
x = np.linspace(-5, 5, n+1)
y = np.zeros(n+1)

for i in range(len(x)):
    y[i] = H(x[i])

print(x, type(x))
print(y, type(y))

plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def H_loop(x):
    r = np.zeros(len(x))
    for i in range(len(x)):
        r[i] = H(x[i])
    return r

n = 5
x = np.linspace(-5, 5, n+1)
y = H_loop(x)

print(x, type(x))
print(y, type(y))

plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

Hv = np.vectorize(H)

def Hv(x):
    return np.where(x<0, 0.0, 1.0)

n = 5
x = np.linspace(-5, 5, n+1)
y = Hv(x)
print(y)

plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    if x <= -3:
        return -3-x
    elif -3 < x <= 0:
        return x + 3
    elif 0 < x <= 3:
        return 3 - 2*x
    else:
        return 0.5*x - 4.5

# print(f(-5), f(-1.5), f(1), f(4))

fv = np.vectorize(f)

n = 20
x = np.linspace(-10, 10, n+1)
y = fv(x)

plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    if x < 0.707:
        return x**2
    else:
        return 1.5-(x-1.414)**2

# print(f(-5), f(-1.5), f(1), f(4))

fv = np.vectorize(f)

n = 50
x = np.linspace(-2, 2, n+1)
y = fv(x)

plt.plot(x, y)
plt.show()
```
