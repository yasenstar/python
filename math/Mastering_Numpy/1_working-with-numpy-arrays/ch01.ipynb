### 1.4.3 Matrix Multiplication

```python
import numpy as np
x = np.array([
    [1, 0, 4],
    [3, 3, 1]
])
y = np.array([
    [2, 5],
    [1, 1],
    [3, 2]
])
print(x.dot(y))
```
