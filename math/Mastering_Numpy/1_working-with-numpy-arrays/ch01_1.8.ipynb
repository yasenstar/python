## 1.8 Indexing, Slicing, Reshaping, Resizing, and Broadcasting

### 1.8.1 Indexing

```python

x = ["China", "France", "Germeny", "Belgium"]
x[2]
```

```python
x = ('USA', 3, "France", 4)
x[1]
```

```python
# Indexing - 1) Field Access
import numpy as np
x = np.arange(10)
print(x)
print(f"x[5]={x[5]}, x[-2]={x[-2]}, x[2:8]={x[2:8]}")
print(f"x[:]={x[:]}, x[2:8:2]={x[2:8:2]}")
```

### 1.8.2 Slicing

```python
# Slice the n-dimensional array as [slicing in x-axis, slicing in y-axis]
x = np.reshape(np.arange(16),(4,4))
print(f"x=\n{x}\nx[1:3]=\n{x[1:3]}\nx[:,1:3]=\n{x[:,1:3]}\nx[1:3,1:3]=\n{x[1:3,1:3]}")
```

```python
x = np.reshape(np.arange(16),(4,4))
print(f"x=\n{x}")
print(f"x[[0,1,2,3],[0,1,3,2]]=\n{x[[0,1,2,3],[0,1,3,2]]}")
```

### 1.8.4 Resizing

```python
np.arange(16).reshape(4,4)
print(f"reshaped x =\n{x}")
print(f"resized x in (2,2) =\n{np.resize(x,(2,2))}")
print(f"resized x in (6,6) =\n{np.resize(x,(6,6))}")
```

### 1.8.5 Broadcasting

```python
x = np.arange(16).reshape(4,4)
y = np.arange(6).reshape(2,3)
print(f"x + y =\n{x+y}")
```

```python
print(f"x * y =\n{x*y}")
```

```python
x = np.arange(16).reshape(4,4)
y = np.arange(4)
print(f"x * y =\n{x*y}")
```

```python
x = np.arange(4).reshape(2,2)
y = np.arange(2).reshape(1,2)
print(f"x * y =\n{x*y}")
```
